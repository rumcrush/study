# 데이터 필터

```python
#연령대 데이터 내용 파악
df1['연령대'].value_counts()
```

```python
#연령대 중에서도 40대 데이터만 추출
#시리즈로 만들어둔 연령대가 40대인 데이터가 cond1에 담김
cond1 = (df1['연령대'] == '40대')
```

```python
#loc를 사용해서 로우 확인
#cond1에 있는 시리즈에 있는 로우를 모두 가져와서 보여줌
df1.loc[cond1]
```

```python
#개수 확인
df1.loc[cond1].shape
```

```python
#40대 제외 항목으로 추출
cond2 = (df1['연령대'] != '40대')
```

```python
df1.loc[cond2]
```

```python
#cond1에 반대되는 조건만 찾기 
df1.loc[~cond1]
```

```python
#연령대가 40대이며 성별이 남 인 사람들의 데이터 
cond1 = (df1['연령대'] == '40대')
cond2 = (df1['성별'] == '남')
df1.loc[cond1 & cond2]
```

```python
#연령대가 40대이거나 성별이 남 인 사람들의 데이터 
df1.loc[cond1 | cond2]
```

# 연습

```python
# 30대 여성의 데이터 추출
```

```python
cond1 = (df1['연령대'] == '30대이하')
cond2 = (df1['성별'] == '여')
```

```python
cond3=df1[cond1 & cond2]
```

```python
#그 중에서 구매금액이 높은 순서대로 정렬 
cond3.sort_values(by='구매금액',ascending=False)

#또는 이렇게 쓴다 
df1[cond1 & cond2].sort_values(by='구매금액', ascending=False)
```

```python
#데이터 프레임 확인 
cond3.shape
```
